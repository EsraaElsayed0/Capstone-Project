```python
import os
import glob
import pandas as pd

def read_txt(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()

def load_text_documents(directory):
    documents = []
    for file in glob.glob(os.path.join(directory, '**', '*.txt'), recursive=True):
        file_name = os.path.basename(file)
        folder_name = os.path.basename(os.path.dirname(file))
        text_content = read_txt(file)
        documents.append({'folder': folder_name, 'filename': file_name, 'text': text_content})
    return documents
```

```python
# Load all text documents from the parent directory
parent_directory = '/content/sanad-dataset'
documents = load_text_documents(parent_directory)

# Create a DataFrame from the loaded documents
df = pd.DataFrame(documents)

# Save the DataFrame to a CSV file
df.to_csv('sanad_dataset.csv', index=False)
```

```python
import re

# Preprocessing function for Arabic text
def preprocess_text_arabic(text):
    # Normalize Arabic text
    text = re.sub(r'[^\u0621-\u064A\s]', '', text)  # Keep only Arabic characters and spaces
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = re.sub(r'\b\w{1,2}\b', '', text)  # Remove short words (1-2 characters)
    text = re.sub(r'#\d{3,}', '', text)  # Remove hashtags with numbers
    text = text.strip()  # Strip leading and trailing spaces
    return text

# Example usage with your DataFrame
df['processed_text'] = df['text'].apply(preprocess_text_arabic)
```

```python
import torch
import numpy as np
from transformers import AutoTokenizer, AutoModel
from tqdm import tqdm
```

```python
tokenizer = AutoTokenizer.from_pretrained("asafaya/bert-base-arabic")
model = AutoModel.from_pretrained("asafaya/bert-base-arabic")
```

```python
def embed_documents(text):
    encoded_text = []

    # Use tqdm for a progress bar
    for chunk in tqdm(text, desc="Embedding Chunks"):
        inputs = tokenizer(chunk, return_tensors='pt', padding=True, truncation=True, max_length=512)
        with torch.no_grad():
            embeddings = model(**inputs).last_hidden_state.mean(dim=1).squeeze().numpy()
        encoded_text.append(embeddings)

    return np.array(encoded_text)
```

```python
import chromadb
from chromadb.config import Settings

# Initialize ChromaDB client without chroma_dir
client = chromadb.Client(Settings())

# Create or get a collection
collection = client.create_collection("document_embeddings")
```

```python
batch_size = 20000  # Set batch size below the limit
for i in range(0, len(df['processed_text']), batch_size):
    batch_ids = [str(j) for j in range(i, min(i + batch_size, len(df['processed_text'])))]
    batch_embeddings = embeddings[i: i + batch_size].tolist()
    batch_metadatas = [{"text": text} for text in df['processed_text'][i: i + batch_size]]
    collection.add(
        ids=batch_ids,
        embeddings=batch_embeddings,
        metadatas=batch_metadatas
    )
```

```python
def embed_query(query):
    """
    Embeds a single query using the BERT-based Arabic model.

    Args:
        query (str): The query to embed.

    Returns:
        np.ndarray: The embedding vector with shape (768,).
    """
    inputs = tokenizer(
        query,
        return_tensors='pt',
        padding=True,
        truncation=True,
        max_length=512
    )
    with torch.no_grad():
        outputs = model(**inputs)
        query_embedding = outputs.last_hidden_state.mean(dim=1).squeeze().numpy()
    return query_embedding
```

```python
# Example query
query ="الشارقة الخليج"
# Embed the query
query_embedding = embed_query(query)
print(f"Query embedding shape: {query_embedding.shape}")  # Should be (768,)
```

```python
# Ensure the query embedding is a 2D list
query_embedding_list = [query_embedding.tolist()]  # Shape: (1, 768)

# Perform the similarity search
results = collection.query(
    query_embeddings=query_embedding_list,
    n_results=5  # Number of top documents to retrieve
)

# Inspect the results
print(results)
```

```python
# Structure the results manually
structured_results = {
    'ids': results.get('ids', []),
    'distances': results.get('distances', []),
    'metadatas': results.get('metadatas', []),
    'embeddings': results.get('embeddings', None),
    'documents': results.get('documents', []),
    'uris': results.get('uris', None),
    'data': results.get('data', None),
    'included': ['metadatas', 'documents', 'distances']
}

# Print structured results
for i, metadata in enumerate(structured_results['metadatas'][0]):
    print(f"Document ID: {structured_results['ids'][0][i]}")
    print(f"Distance: {structured_results['distances'][0][i]}")
    print(f"Text: {metadata['text']}\n")
```

```python
def generate_response(query, retrieved_docs, llm):
    """
    Generates a response based on the query and retrieved document chunks.

    Args:
        query (str): The user's query.
        retrieved_docs (list of dict): Retrieved document chunks.
        llm: The language model to generate responses.

    Returns:
        str: The generated response.
    """
    # Combine the content from retrieved documents
    context = " ".join([doc['text'] for doc in retrieved_docs])

    # Prepare the input for the generative model
    input_data = {
        "question": query,
        "context": context
    }

    # Define the prompt template
    template = """
    You are an intelligent assistant created to use the information retrieved from the knowledge base to generate accurate answers.

    Question:
    {question}

    Context:
    {context}
    """

    rag_prompt = PromptTemplate.from_template(template)

    # Create the chain to generate response
    rag_chain = (
        {"context": RunnablePassthrough(), "question": RunnablePassthrough()}
        | rag_prompt
        | llm
        | StrOutputParser()
    )

    # Generate the response
    response = rag_chain.invoke(input_data)
    return response

def rag_pipeline(query, collection, llm, top_k=5):
    """
    The complete RAG pipeline: retrieves relevant documents and generates a response.

    Args:
        query (str): The user's query.
        collection: The ChromaDB collection object.
        llm: The language model for generating responses.
        top_k (int): Number of top documents to retrieve.

    Returns:
        str: The generated response.
    """
    # Embed the query using BERT
    query_embedding = embed_query(query)
    query_embedding = [query_embedding.tolist()]  # Wrap the embedding in another list

    # Step 1: Retrieve relevant documents using ChromaDB
    results = collection.query(query_embeddings=query_embedding, n_results=top_k)
    retrieved_docs = results['metadatas'][0]  # Retrieve the document metadata (texts)
    print(retrieved_docs)
    # Step 2: Generate a response based on retrieved documents
    response = generate_response(query, retrieved_docs, llm)
    return response

# Example usage
query = "الشارقة الخليج"
response = rag_pipeline(query, collection, llm, top_k=5)
print(response)
```

```python
import gradio as gr

def gradio_interface(query):
    # Step 1: Embed the query and retrieve documents
    query_embedding = embed_query(query)
    query_embedding = [query_embedding.tolist()]
    results = collection.query(query_embeddings=query_embedding, n_results=5)  # Adjust 'n_results' as needed

    if not results or not results['metadatas']:
        return "No documents found."

    # Step 2: Generate a response based on the retrieved documents
    retrieved_docs = results['metadatas'][0]  # Extract the metadata (which contains text)
    response = generate_response(query, retrieved_docs, llm)  # Use the generative model to produce a response

    # Step 3: Create a formatted string for display
    display_text = f"**Generated Response**:\n{response}\n\n---\n\n"

    for i, doc in enumerate(retrieved_docs):
        display_text += f"**Document {i+1}**:\n"
        display_text += f"**Text**: {doc['text']}\n"
        display_text += f"**Distance**: {results['distances'][0][i]:.4f}\n\n---\n\n"

    return display_text

# Define the Gradio interface
iface = gr.Interface(
    fn=gradio_interface,
    inputs=gr.Textbox(lines=2, placeholder="Enter your query here...", label="Query"),
    outputs=gr.Markdown(label="Retrieved Documents and Generated Response"),
    title="Arabic Text Retrieval and Generation System",
    description="Enter an Arabic query to retrieve the most relevant documents from the database and generate a response.",
    examples=[
        ["سامسونغ الخليج للإلكترونيات"],
        ["جهاز غالكسي نوت 2"],
        ["مركز الاتصالات"]
    ]
)

# Launch the interface
iface.launch()
```
