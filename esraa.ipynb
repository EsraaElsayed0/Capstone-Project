```python
import os
import glob
import pandas as pd

def read_txt(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()

def load_text_documents(directory):
    documents = []
    for file in glob.glob(os.path.join(directory, '**', '*.txt'), recursive=True):
        file_name = os.path.basename(file)
        folder_name = os.path.basename(os.path.dirname(file))
        text_content = read_txt(file)
        documents.append({'folder': folder_name, 'filename': file_name, 'text': text_content})
    return documents
```

```python
# Load all text documents from the parent directory
parent_directory = '/content/sanad-dataset'
documents = load_text_documents(parent_directory)

# Create a DataFrame from the loaded documents
df = pd.DataFrame(documents)

# Save the DataFrame to a CSV file
df.to_csv('sanad_dataset.csv', index=False)
```

```python
import re
# Preprocessing function
def preprocess_text(text):
    text = text.lower()  # Lowercase the text
    text = re.sub(r'\b\w{1,2}\b', '', text)  # Remove short words
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    return text.strip()

# Apply preprocessing
df['processed_text'] = df['text'].apply(preprocess_text)
```

```python
import torch
import numpy as np
from transformers import AutoTokenizer, AutoModel
from tqdm import tqdm

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
model = AutoModel.from_pretrained("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
def embed_documents(text):
    encoded_text = []

    # Use tqdm for a progress bar
    for chunk in tqdm(text, desc="Embedding Chunks"):
        inputs = tokenizer(chunk, return_tensors='pt', padding=True, truncation=True, max_length=512)
        with torch.no_grad():
            embeddings = model(**inputs).last_hidden_state.mean(dim=1).squeeze().numpy()
        encoded_text.append(embeddings)

    return np.array(encoded_text)
```

```python
# Assuming df['chunks'] contains the text chunks
embeddings = embed_documents(df['text'])  # Calculate embeddings

# Create a new DataFrame from the embeddings
embeddings_df = pd.DataFrame(embeddings)

# Concatenate the original DataFrame with the embeddings DataFrame
df = pd.concat([df, embeddings_df], axis=1)

# Display the DataFrame with the new embeddings columns
df.head()
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the T5 tokenizer and model
tokenizer = GPT2LMHeadModel.from_pretrained('gpt2')
generator_model = GPT2Tokenizer.from_pretrained('gpt2')
```

```python
def generate_response(query, retrieved_docs):
    """
    Generates a response based on the query and retrieved document chunks.

    Args:
        query (str): The user's query.
        retrieved_docs (list of Document): Retrieved document chunks.

    Returns:
        str: The generated response.
    """
    # Combine the content from retrieved documents
    context = " ".join([doc.page_content for doc in retrieved_docs])

    # Prepare the input for the T5 model
    input_text = f"question: {query} context: {context}"

    # Tokenize the input
    input_ids = tokenizer.encode(input_text, return_tensors='pt', truncation=True)

    # Generate the response
    output_ids = generator_model.generate(
        input_ids,
        max_length=512,
        num_return_sequences=1,
        num_beams=4,
        early_stopping=True
    )

    # Decode the generated tokens to text
    response = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return response

def rag_pipeline(query, retriever, generator_model, tokenizer, top_k=5):
    """
    The complete RAG pipeline: retrieves relevant documents and generates a response.

    Args:
        query (str): The user's query.
        retriever: The retriever object.
        generator_model: The generative model.
        tokenizer: The tokenizer for the generative model.
        top_k (int): Number of top documents to retrieve.

    Returns:
        str: The generated response.
    """
    query_embedding = embed_documents([query])[0]
    # Step 1: Retrieve relevant documents
    retrieved_docs = retriever.get_relevant_documents(query_embedding)[:top_k]

    # Step 2: Generate a response based on retrieved documents
    response = generate_response(query, retrieved_docs)

    return response
```
