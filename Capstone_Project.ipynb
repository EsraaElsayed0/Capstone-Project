```python
import os
import glob
import pandas as pd

def read_txt(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()

def load_text_documents(directory):
    documents = []
    for file in glob.glob(os.path.join(directory, '**', '*.txt'), recursive=True):
        file_name = os.path.basename(file)
        folder_name = os.path.basename(os.path.dirname(file))
        text_content = read_txt(file)
        documents.append({'folder': folder_name, 'filename': file_name, 'text': text_content})
    return documents
```

```python
# Load all text documents from the parent directory
parent_directory = '/content/sanad-dataset'
documents = load_text_documents(parent_directory)

# Create a DataFrame from the loaded documents
df = pd.DataFrame(documents)

# Save the DataFrame to a CSV file
df.to_csv('sanad_dataset.csv', index=False)
```

```python
import re
# Preprocessing function
def preprocess_text(text):
    text = text.lower()  # Lowercase the text
    text = re.sub(r'\b\w{1,2}\b', '', text)  # Remove short words
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    return text.strip()

# Apply preprocessing
df['processed_text'] = df['text'].apply(preprocess_text)
```

```python
import torch
import numpy as np
from transformers import AutoTokenizer, AutoModel
from tqdm import tqdm

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
model = AutoModel.from_pretrained("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
```

```python
def embed_documents(text):
    encoded_text = []

    # Use tqdm for a progress bar
    for chunk in tqdm(text, desc="Embedding Chunks"):
        inputs = tokenizer(chunk, return_tensors='pt', padding=True, truncation=True, max_length=512)
        with torch.no_grad():
            embeddings = model(**inputs).last_hidden_state.mean(dim=1).squeeze().numpy()
        encoded_text.append(embeddings)

    return np.array(encoded_text)
```

```python
# Assuming df['chunks'] contains the text chunks
embeddings = embed_documents(df['processed_text'])  # Calculate embeddings
```

```python
import gradio as gr
import faiss
import numpy as np
import torch
from transformers import AutoTokenizer, AutoModel, GPT2LMHeadModel, GPT2Tokenizer
from langchain.vectorstores import FAISS
from langchain.schema import Document

# Load your FAISS index and texts
index_path = '/content/Database_store/MyDrive/vector_database.index'  # Adjust the path if necessary
texts_path = '/content/Database_store/MyDrive/texts.txt'

# Load the FAISS index
try:
    index = faiss.read_index(index_path)
except Exception as e:
    raise ValueError(f"Error loading FAISS index: {e}")

# Load the texts
try:
    with open(texts_path, 'r') as f:
        texts = f.read().splitlines()
except Exception as e:
    raise ValueError(f"Error loading texts: {e}")

# Load the Hugging Face model and tokenizer for embeddings
try:
    tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
    model = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
except Exception as e:
    raise ValueError(f"Error loading embedding model: {e}")

# Load the GPT-2 model and tokenizer for generating responses
try:
    gpt2_tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    gpt2_model = GPT2LMHeadModel.from_pretrained('gpt2')
except Exception as e:
    raise ValueError(f"Error loading GPT-2 model: {e}")

# Define an embedding function
def embed_query(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        outputs = model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1).cpu().numpy()
    return embeddings

# Create the FAISS vectorstore
try:
    faiss_db = FAISS(embed_query, index, texts, {})
except Exception as e:
    raise ValueError(f"Error creating FAISS vectorstore: {e}")

# Create the retriever from the FAISS vectorstore
try:
    retriever = faiss_db.as_retriever()
except Exception as e:
    raise ValueError(f"Error creating retriever: {e}")

def getting_answers(question, history=[]):
    try:
        # Retrieve relevant documents using the retriever
        docs = retriever.get_relevant_documents(question)

        if not docs:
            return "No relevant documents found.", "No similar texts available."

        # Combine the retrieved documents into a single context
        similar_texts = "\n".join([doc.page_content for doc in docs])

        # Prepare the prompt for GPT-2 by combining the query with the context
        prompt = f"Context:\n{similar_texts}\n\nQuestion: {question}\nAnswer:"

        # Tokenize and generate the response using GPT-2
        inputs = gpt2_tokenizer(prompt, return_tensors="pt")
        outputs = gpt2_model.generate(inputs['input_ids'], max_length=200)

        # Decode the generated text
        answer = gpt2_tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Print statements for debugging
        print(f"Answer: {answer}")
        print(f"Similar Texts: {similar_texts}")

        # Return the answer and the similar texts
        return answer, similar_texts
    except Exception as e:
        # Print the error for debugging
        print(f"Error: {str(e)}")
        # Return an error message and an empty string
        return f"An error occurred while generating the answer: {str(e)}", ""

# Define the Gradio interface
demo = gr.Interface(
    fn=getting_answers,
    inputs=["text"],
    outputs=["textbox", "textbox"],
    title="AI Q&A with Context Retrieval",
    description="Enter a question, and the AI will provide an answer along with similar texts from the dataset."
)

# Launch the interface
demo.launch()
```
